Інтерполяційний багаточлен Лагранжа

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, expand
```

```python
import numpy as np

# Табличні дані
x = np.array([-2., -1., 0., 1.], dtype=float)
y = np.array([-7., 4., 1., 2.], dtype=float)

# Точки для обчислення
x1, x2, x3, x4 = -3, -1.5, 0.5, 1.5

# Інтерполяція Лагранжа
def lagrange_interpolation(x, y, x_test):

    n = len(x)
    p = np.zeros(n)

    for i in range(n):

        p_i = 1

        for j in range(n):

            if i != j:
                p_i *= (x_test - x[j]) / (x[i] - x[j])

        p[i] = p_i

    return np.dot(y, p)

# Обчислення значень
f1 = lagrange_interpolation(x, y, x1)
print("Значення функції у точці x1 =", round(f1, 4))

f2 = lagrange_interpolation(x, y, x2)
print("Значення функції у точці x2 =", round(f2, 4))

f3 = lagrange_interpolation(x, y, x3)
print("Значення функції у точці x3 =", round(f3, 4))

f4 = lagrange_interpolation(x, y, x4)
print("Значення функції у точці x4 =", round(f4, 4))
```

```python
xnew=np.linspace(np.min(x),np.max(x),100) #точки, за якими будуємо графік
ynew=[lagrange_interpolation(x,y,i) for i in xnew]

plt.plot(x,y,'o',xnew,ynew) #будуємо графік функції Лагранжа

plt.plot(x1,f1,'ro')
plt.plot(x2,f2,'ro')
plt.plot(x3,f3,'ro')
plt.plot(x4,f4,'ro')


plt.xlabel('x')
plt.ylabel('y')

plt.title('Lagrange Polynomial_1')
plt.grid(True)
plt.show()
```

```python
from scipy.interpolate import lagrange

#Перевірка
f = lagrange(x, y)
fig = plt.figure(figsize = (7,5))
plt.plot(xnew, f(xnew), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial_2')

plt.plot(x1,f1,'ro')
plt.plot(x2,f2,'ro')
plt.plot(x3,f3,'ro')
plt.plot(x4,f4,'ro')

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Приклад

```python
from scipy.interpolate import lagrange

# Таблиця значень функції
x = [-2, -1, 0, 1]
f = [-7, 4, 1, 2]

d = len(x)

a = []
L = []

# Формування коефіцієнтів і базисних поліномів
for j in range(d):

    p = 1
    s = ''

    for i in range(d):

        if i == j:
            continue

        else:
            p = p * (x[j] - x[i])
            s = s + '(x-' + str(x[i]) + ')'

    k = f[j] / p

    a.append(k)
    L.append(s)

# Формування полінома
polynom_string = 'L(x) = '

for i in range(d):
    polynom_string = polynom_string + str(a[i]) + '*' + L[i] + '+'

polynom_string = polynom_string[:-1].replace('+-', '-')

# Виведення
print('Інтерполяційний поліном Лагранжа')
print(polynom_string)
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Таблиця значень функції (варіант №1)
x = np.array([-2, -1, 0, 1], dtype=float)
y = np.array([-7, 4, 1, 2], dtype=float)

# Інтерполяційний поліном Лагранжа
p = lagrange(x, y)

# Виведення полінома
print("Інтерполяційний поліном Лагранжа:")
print(p)

# Точки для побудови графіка
z = np.linspace(np.min(x), np.max(x), 100)

# Графік
plt.plot(z, p(z), '-')
plt.plot(x, y, 'o')

plt.legend(['L(x)', 'Табличні точки'])
plt.grid(True)

plt.xlabel('x')
plt.ylabel('y')

plt.title('Інтерполяція поліномом Лагранжа')

plt.show()
```
