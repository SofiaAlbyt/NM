```python
import numpy as np
import matplotlib.pyplot as plt

# Межі графіка
x_min, x_max = -3, 3
y_min, y_max = -3, 3
step = 0.01

# Сітка значень
x, y = np.meshgrid(
    np.arange(x_min, x_max, step),
    np.arange(y_min, y_max, step)
)

# Система рівнянь
eq1 = np.sin(x + 1) - y - 1.2
eq2 = 2 * x + np.cos(y) - 2

# Побудова графіка
fig, ax = plt.subplots(figsize=(8, 8))

# Перше рівняння
ax.contour(x, y, eq1, levels=[0], colors='red')

# Друге рівняння
ax.contour(x, y, eq2, levels=[0], colors='blue')

# Оформлення
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_title('Графічне розв’язання системи рівнянь')

# Осі
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Сітка
plt.grid(True)

# Показ графіка
plt.show()
```

```python
# Розв'язок системи рівнянь методом простої ітерації

import numpy as np
from scipy import optimize
import math

x0, y0 = 0.5, 0.5
eps = 0.001

# Функції для ітерацій
def f_x(y):
    return (2 - math.cos(y)) / 2

def f_y(x):
    return math.sin(x + 1) - 1.2

# Метод простої ітерації
def simple_iteration(x, y, eps, max_iter=1000):

    xn = x
    yn = y

    xn1 = f_x(yn)
    yn1 = f_y(xn)

    n = 1

    while (abs(xn1 - xn) >= eps) or (abs(yn1 - yn) >= eps):

        xn = xn1
        yn = yn1

        xn1 = f_x(yn)
        yn1 = f_y(xn)

        n += 1

        if n > max_iter:
            raise RuntimeError("Не зійшлось за max_iter")

    print('Simple iteration:')
    print(f'x = {xn1:.5f}')
    print(f'y = {yn1:.5f}')
    print(f'iterations = {n}')

    return xn1, yn1, n

# Виклик методу
x_iter, y_iter, n_iter = simple_iteration(x0, y0, eps)

# Перевірка через scipy
def system(p):
    x, y = p
    return [
        math.sin(x + 1) - y - 1.2,
        2 * x + math.cos(y) - 2
    ]

sol = optimize.root(system, [x0, y0], method='hybr')

print("\nCheck (scipy):")
print(sol.x)
```
