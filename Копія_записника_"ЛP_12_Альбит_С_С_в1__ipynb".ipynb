Завдання 3. Визначити інтеграм методом трапецій

```python
# Обчислення інтегралу методом трапецій
from scipy import integrate
import numpy as np

# Задаємо функцію, яку необхідно інтегрувати
def f(x):
    return 1 / np.sqrt(x**2 + 2.3)

# Задаємо межі інтегрування та кількість розбиттів
a, b, n = 0.32, 0.66, 20

# Метод трапецій
def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    s = 0.5 * (f(a) + f(b))
    x = a

    for i in range(1, n):
        x += h
        s += f(x)

    integral = h * s
    return integral

# Обчислення
integral = trapezoidal_rule(f, a, b, n)
print("Trapezoidal method result:", round(integral, 5))

# Перевірка через SciPy
v, err = integrate.quad(f, a, b)
print("Check with SciPy quad method:", round(v, 5))
```

Завдання 1. Метод прямокутників (правих, лівих і середніх)

```python
# Обчислення інтеграла методом прямокутників
from scipy import integrate
import numpy as np

a, b, n = 0.8, 1.6, 10

h = (b - a) / n  # Крок

def f1(x):
    return 1 / np.sqrt(2 * x**2 + 1)

# Метод лівих прямокутників
def left_rec(f1, a, b):
    s = 0
    for i in range(0, n):
        s += f1(a + i * h)
    return s * h

print("Left rectangle:", round(left_rec(f1, a, b), 5))

# Метод правих прямокутників
def right_rec(f1, a, b):
    s = 0
    for i in range(1, n + 1):
        s += f1(a + i * h)
    return s * h

print("Right rectangle:", round(right_rec(f1, a, b), 5))

# Метод середніх прямокутників
def aver_rec(f1, a, b):
    s = 0
    for i in range(0, n):
        s += f1(a + (i + 0.5) * h)
    return s * h

print("Average rectangle:", round(aver_rec(f1, a, b), 5))

# Перевірка
v, err = integrate.quad(f1, a, b)
print("Check with SciPy quad method:", round(v, 5))

```

Завдання 2. Метод Симпсона

```python
# Обчислення інтеграла методом Симпсона
from scipy import integrate
import numpy as np

# Функція
def f(x):
    return np.log10(x + 2) / x

# Межі та кількість розбиттів
a, b, n = 1.2, 2, 8

h = (b - a) / n

# Метод Сімпсона
def simpson_rule(f, a, b, n):
    integr = f(a) + f(b)

    for i in range(1, n):
        k = a + i * h
        if i % 2 == 0:
            integr += 2 * f(k)
        else:
            integr += 4 * f(k)

    integr *= h / 3
    return integr

# Результат
print("Simpson method:", round(simpson_rule(f, a, b, n), 6))

# Перевірка
v, err = integrate.quad(f, a, b)
print("Check with SciPy quad:", round(v, 6))
```
