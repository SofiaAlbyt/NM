1. Метод ділення навпіл.

```python
import numpy as np
import pandas as pd

# Визначаємо функцію
def f(x):
    y = 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
    return y

# Визначаємо границі відрізку
a, b = 0, 1

# Визначаємо точність розв'язку
eps = 0.001

# Поділяємо відрізок навпіл
c = (a + b) / 2

# Зберігаємо результати обчислень
data_A = [a]
data_B = [b]
data_C = [c]

data_F_A = [f(a)]
data_F_B = [f(b)]
data_F_C = [f(c)]

# Метод половинного ділення
while np.abs(f(c)) > eps:

    # Звужуємо відрізок
    if np.sign(f(c)) == np.sign(f(a)):
        a = c
    else:
        b = c

    # Нова середина
    c = (a + b) / 2

    # Зберігаємо результати
    data_A.append(a)
    data_B.append(b)
    data_C.append(c)

    data_F_A.append(f(a))
    data_F_B.append(f(b))
    data_F_C.append(f(c))

# Відображаємо результати
print('Результати обчислень')
print()

data = pd.DataFrame({
    'a': data_A,
    'F(a)': data_F_A,
    'c': data_C,
    'F(c)': data_F_C,
    'b': data_B,
    'F(b)': data_F_B
})

print(data)

# Відображаємо корінь
print()
print('В точці c =', c, ', f(c) =', f(c))
```

3. Метод хорд

```python
import numpy as np
import pandas as pd

# Визначаємо функцію
def f(x):
    y = 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
    return y

# Межі відрізку
a, b = 0, 1

# Точність
eps = 0.001

# Точка перетину хорди з віссю Ox
c = (f(a) * b - a * f(b)) / (f(a) - f(b))

# Збереження результатів
data_A = [a]
data_B = [b]
data_C = [c]

data_F_A = [f(a)]
data_F_B = [f(b)]
data_F_C = [f(c)]

# Метод хорд
while np.abs(f(c)) > eps:

    # Звуження відрізка
    if np.sign(f(c)) == np.sign(f(a)):
        a = c
    else:
        b = c

    # Нова точка хорди
    c = (f(a) * b - a * f(b)) / (f(a) - f(b))

    # Збереження результатів
    data_A.append(a)
    data_B.append(b)
    data_C.append(c)

    data_F_A.append(f(a))
    data_F_B.append(f(b))
    data_F_C.append(f(c))

# Виведення результатів
print('Результати обчислень')
print()

data = pd.DataFrame({
    'a': data_A,
    'F(a)': data_F_A,
    'c': data_C,
    'F(c)': data_F_C,
    'b': data_B,
    'F(b)': data_F_B
})

print(data)

# Виведення кореня
print()
print('В точці c =', c, ', f(c) =', f(c))
```

4. Метод Ньютона (дотичних)

```python
import numpy as np
import pandas as pd

# Визначаємо функцію
def f(x):
    y = 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
    return y

# Визначаємо межі відрізку
a, b = 0, 1

# Визначаємо першу похідну
def df(x):
    y = 36*x**3 + 24*x**2 + 3*x + 2
    return y

# Початкове наближення
c = 0.7

# Точність розв'язку
eps = 0.001

# Зберігаємо результати обчислень
data_C = [c]
data_F_C = [f(c)]

# Метод Ньютона
while np.abs(f(c)) > eps:
    c -= f(c) / df(c)

    data_C.append(c)
    data_F_C.append(f(c))

# Відображаємо результати
print('Результати обчислень')
print()

data = pd.DataFrame({
    'c': data_C,
    'F(c)': data_F_C
})

print(data)

# Відображаємо наближений корінь
print()
print('В точці c =', c, ', f(c) =', f(c))
```

```python
import numpy as np
import scipy.optimize as sp

# Визначаємо функцію
def f(x):
    y = 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
    return y

# Обчислення похідної
def df(x):
    y = 36*x**3 + 24*x**2 + 3*x + 2
    return y

# Початкові наближення
x0, x1, x2, x3, x4 = -1.4, 1.4, -3.2, 3.2, 0.5

# Метод Ньютона
x_0 = sp.newton(f, x0, df, tol=1.e-3, maxiter=50)
x_1 = sp.newton(f, x1, df, tol=1.e-3, maxiter=50)

x_2 = sp.newton(f, x2, df, tol=1.e-3, maxiter=50)
x_3 = sp.newton(f, x3, df, tol=1.e-3, maxiter=50)

x_4 = sp.newton(f, x4, df, tol=1.e-3, maxiter=50)

# Виведення результатів
print('x_0 =', x_0, 'f(x_0) =', f(x_0))
print('x_1 =', x_1, 'f(x_1) =', f(x_1))
print('x_2 =', x_2, 'f(x_2) =', f(x_2))
print('x_3 =', x_3, 'f(x_3) =', f(x_3))
print('x_4 =', x_4, 'f(x_4) =', f(x_4))
```

```python
# Лабораторна робота 2
# Відокремлення коренів

import numpy as np
import numdifftools as nd

# Функція
def f(x):
    return 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10

# Відокремлення коренів
def find_segments():

    search_range = np.arange(-5, 5, 1)

    a = None
    previous_x = None
    current_x = None

    segments = []

    for x in search_range:

        x = round(x, 4)

        current_x = f(x)

        if previous_x is not None and previous_x * current_x < 0:
            segments.append((a, x))

        a = x
        previous_x = current_x

    return segments

# Пошук відрізків
segments = find_segments()

# Виведення результатів
for a, b in segments:
    print(f'Found segment: [{a}, {b}]')
```

```python
# Лабораторна робота
# Метод половинного ділення
import numpy as np
def f(x):
    return 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
def rec(a, b, eps):
    while np.abs(a - b) > eps:
        if f(a) * f((a + b) / 2) < 0:
            b = (a + b) / 2
        else:
            a = (a + b) / 2
        x = (a + b) / 2
    print('x =', round(x, 5), '- Half division method')
rec(0, 1, 0.001)
```

```python
import numpy as np
import numdifftools as nd
def f(x):
    return 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
def hord(a, b, eps):
    derivative_f = nd.Derivative(f, n=2)
    if f(a) * derivative_f(a) > 0:
        x0 = a
        xi = b
    else:
        x0 = b
        xi = a
    xi_1 = xi - (xi - x0) * f(xi) / (f(xi) - f(x0))
    while np.abs(xi_1 - xi) > eps:
        xi = xi_1
        xi_1 = xi - (xi - x0) * f(xi) / (f(xi) - f(x0))
    print('x =', round(xi_1, 5), '- Chord method')
hord(0, 1, 0.001)
```

```python
# Лабораторна робота 3
# Метод Ньютона (дотичних)

import numpy as np
import numdifftools as nd

# Функція
def f(x):
    return 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10

# Метод Ньютона
def nuton(a, b, eps, f):

    # Перша похідна
    df1 = nd.Derivative(f, n=1)

    # Друга похідна
    df2 = nd.Derivative(f, n=2)

    # Вибір початкової точки
    xi = b if f(b) * df2(b) > 0 else a

    # Перше наближення
    xi_1 = xi - f(xi) / df1(xi)

    # Ітераційний процес
    while np.abs(xi_1 - xi) > eps:
        xi = xi_1
        d = df1(xi)
        if d == 0:
            raise ZeroDivisionError(
                "Похідна дорівнює нулю — зупинка.")
        xi_1 = xi - f(xi) / d
    print('Метод Ньютона, x =', round(float(xi_1), 4))
nuton(0, 1, 0.001, f)
```

```python
# Комбінований метод

import numpy as np
import numdifftools as nd
def f(x):
    return 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10
def komb(a, b, eps, f):
    df2 = nd.Derivative(f, n=2)(a)
    if f(a) * df2 > 0:
        a0, b0 = a, b
    else:
        a0, b0 = b, a
    ai, bi = a0, b0
    while np.abs(ai - bi) > eps:
        ai_1 = ai - f(ai) * (bi - ai) / (f(bi) - f(ai))
        bi_1 = bi - f(bi) / nd.Derivative(f, n=1)(bi)
        ai, bi = ai_1, bi_1
    x = (ai_1 + bi_1) / 2
    print('Комбінований метод, x =', round(float(x), 4))
komb(0, 1, 0.001, f)
```

```python
if __name__ == "__main__":
    a1, b1 = -2, -1
    a2, b2 = 0, 1
    eps = 0.001
    print("Розв'язання на відрізку [-2, -1]")
    nuton(a1, b1, eps, f)
    komb(a1, b1, eps, f)
    print("\nРозв'язання на відрізку [0, 1]")
    nuton(a2, b2, eps, f)
    komb(a2, b2, eps, f)
```

```python
def f(x: float):
    return x**3 + 0.1*x**2 + 0.4*x + 1.2

def fprime(x: float):
    return nd.Derivative(f, n=1)

def phi(x: float):
    # c = -0.25 (забезпечує |phi'(x)|<1 на [-1,0])
    return x - 0.25 * f(x)

def iterate_fixed_point(x0: float, iters: int = 5):
    xs = [x0]
    x = x0
    for _ in range(iters):
        x = phi(x)
        xs.append(x)
    return xs

def error_bound(x_last: float) -> float:
    # μ = min_{x∈[-1,0]} |f'(x)| = 0.4
    mu = 0.4
    return abs(f(x_last)) / mu

```

```python
# Рівняння: f(x) = x³ − 0.1x² + 0.4x + 1.2 = 0
# Ітераційна функція: φ(x) = x - 0.1 * f(x)  (c = -0.25)
# Початкове наближення: x0 = -1
# Кількість ітерацій: 5
# Оцінка похибки: |ξ - x5| ≤ |f(x5)| / μ, де μ = min_{[-3,-1]} |f'(x)| = 0.4

def f(x):
    return x**3 - 0.1*x**2 + 0.4*x + 1.2

def phi(x):
    return x - 0.25 * f(x)   # c = -0.25

def fixed_point(x0=-1.0, iters=5):
    xs = [x0]
    for _ in range(iters):
        xs.append(phi(xs[-1]))
    return xs

# Запуск і вивід результатів
xs = fixed_point(-1.0, 5)

for i, xi in enumerate(xs):
    print(f"x_{i} = {xi:.10f}")

# Похибка методу простої ітерації після 5-ї ітерації
mu = 0.4                      # μ = min_{[-3,-1]} |3x^2 - 1| = 0.4
err_bound = abs(f(xs[-1]))/mu
print(f"\nОцінка похибки: |ξ - x5| ≤ {err_bound:.10f}")
```
